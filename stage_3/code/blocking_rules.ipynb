## Code for blocking rules

### Providing input file locations

```python
table_a = '/u/s/p/sparimi/cs839/stage3/imdb'
table_b = '/u/s/p/sparimi/cs839/stage3/rotton_tom'
candidate_set = '/u/s/p/sparimi/cs839/stage3/candidate_set'
prediction_set = '/u/s/p/sparimi/cs839/stage3/prediction_list'
```

### Performing blocking

```python
import pandas as pd
import py_entitymatching as em

A = em.read_csv_metadata(table_a, key='_id')
B = em.read_csv_metadata(table_b, key='_id')



ob = em.OverlapBlocker()

#Block tables based on movie-name and director attributes
C = ob.block_tables(A, B, 'Name', 'Name', word_level=True, overlap_size=1, rem_stop_words=True, l_output_attrs=['Name','Genre', 'Director'], r_output_attrs=['Name', 'Genre', 'Director'], show_progress=False)
C2 = ob.block_candset(C, 'Director', 'Director', word_level=True, overlap_size=1, show_progress=False)

C2
```

```python
#Save the reduced candidate set 
C2.to_csv(path_or_buf='/u/s/p/sparimi/cs839/stage3/cand_set_after_blocking',columns=['ltable__id', 'rtable__id'], header=['A_id', 'B_id'], index=False, index_label=False)
```
